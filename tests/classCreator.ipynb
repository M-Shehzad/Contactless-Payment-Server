```python
import os
import pandas as pd
import tensorflow as tf
import cv2
import numpy as np
```

```python
model = tf.keras.models.load_model('./0000001')
```

```python
roi_path = 'C:/Users/maxbr/Desktop/development stuff/Final year project/ROI/Right'
```

```python
images = []
file_paths = []
for filename in sorted(os.listdir(roi_path)):
    file_paths.append(filename)
    img = cv2.imread(os.path.join(roi_path,filename))
    images.append(img)
```

```python
img_list =[]

for filename in sorted(os.listdir(roi_path)):
        integer_val = int(filename.split('_')[0])
        img_list.append([filename, integer_val])
```

```python
img_list
```

```python
df = pd.DataFrame(img_list, columns=['filename', 'integer_val'])
```

```python
def predictor_of_image(image):
    np_image = np.array(image.astype('float32')/255.0)
    gray = cv2.cvtColor(np_image, cv2.COLOR_BGR2GRAY)
    np_image = gray.reshape(-1,150,150,1)
    return model.predict(np_image)
```

```python
vectors = []
for img in images:
    vector = predictor_of_image(img)
    vectors.append(vector[0])
```

```python
pandas_vectors = pd.DataFrame(vectors)
```

```python
pandas_vectors
```

```python
#merge to dataframes
df = pd.concat([df,pandas_vectors], axis=1)
```
